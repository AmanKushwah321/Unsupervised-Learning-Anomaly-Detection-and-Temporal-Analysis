###Question 1: What is Dimensionality Reduction? Why is it important in machine learning?

Answer
**Dimensionality Reduction** is the process of reducing the number of input features (variables) in a dataset while preserving as much meaningful information as possible.

**Why Important?**  
- Reduces **computational cost** (faster training).  
- Avoids **curse of dimensionality** (poor performance in high dimensions).  
- Removes **noise and redundant features**.  
- Improves **model accuracy and generalization**.  
- Enables **data visualization** (e.g., 2D/3D plots).  

> *For Aman (77), reducing features in the Pima Diabetes dataset (8 features → 2 via PCA) helps visualize diabetic vs. non-diabetic patterns.*

###Question 3: What is clustering in unsupervised learning? Mention three popular clustering algorithms?

Answer
**Clustering** groups similar data points without labels using patterns in the data.

**Three Popular Algorithms**:  
1. **K-Means** – Partitions data into K clusters by minimizing within-cluster variance.  
2. **Hierarchical Clustering** – Builds a tree of clusters (dendrogram).  
3. **DBSCAN** – Density-based; finds core points and expands clusters. Handles noise.

###Question 4: Explain the concept of anomaly detection and its significance?

**Answer**:  
**Anomaly Detection** identifies rare or unusual data points that deviate significantly from normal behavior.

**Significance**:  
- **Fraud detection** (e.g., credit card misuse).  
- **Network security** (intrusion detection).  
- **Equipment failure prediction**.  
- **Medical diagnosis** (e.g., detecting tumors).

###Question 5: List and briefly describe three types of anomaly detection techniques.**

Answer

| Type | Description |
|------|-----------|
| **Statistical** | Uses mean, std dev; flags points beyond thresholds (e.g., Z-score > 3). |
| **Distance-based** | Flags points far from neighbors (e.g., KNN distance). |
| **Density-based** | Flags points in low-density regions (e.g., LOF – Local Outlier Factor). |

###Question 6: What is time series analysis? Mention two key components of time series data.**
Answer
**Time Series Analysis** studies data points collected at successive time intervals to identify trends, patterns, and forecast future values.
####Two Key Components
1. **Trend** – Long-term increase or decrease.  
2. **Seasonality** – Repeating patterns at fixed intervals (e.g., daily, yearly).

###Question 8: Write Python code to perform K-means clustering on a sample dataset?

```python
# K-Means Clustering Example
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Generate sample data
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=42)

# Apply K-Means
kmeans = KMeans(n_clusters=4, random_state=42)
y_pred = kmeans.fit_predict(X)

# Plot results
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', marker='x', s=200, linewidths=3, label='Centroids')
plt.title("Aman’s K-Means Clustering (K=4)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
print(f"Inertia (Sum of squared distances): {kmeans.inertia_:.2f}")
```

###Question 9: What is inheritance in OOP? Provide a simple example in Python?
Answer
**Inheritance** allows a class (child) to inherit attributes and methods from another class (parent), promoting code reuse.

```python
# Inheritance Example
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def introduce(self):
        print(f"Hi, I am {self.name}, {self.age} years old.")
class Student(Person):  # Inherits from Person
    def __init__(self, name, age, grade):
        super().__init__(name, age)
        self.grade = grade
    def study(self):
        print(f"{self.name} is working in {self.grade}.")

# Usage
aman = Student("Ansh", 27, "HDFC bank")
aman.introduce()  # Inherited method
aman.study()      # Child method
```
